# Quick overview 

As this is a rather small event (68 players) and there was heavy metagaming beforehand, the data will probably not be too useful outside of this tournament. Still, it's interesting to see which teams came up with the best plans to beat each other.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

data = pd.read_csv("datasets/mcv_decks.csv", sep=",")
print(data.describe())
```

```python

# all decks
total_finishes = data['Deck'].value_counts()

# decks with 6 wins or more are most interesting, as they represent a reasonable cutoff point
top_performers = data.loc[data['Wins']>= 6]
tp = top_performers['Deck'].value_counts()

df = pd.DataFrame({'Deck':total_finishes.index, 'Total Finishes': total_finishes.values})
df2 = pd.DataFrame({'Deck':tp.index, 'Top Finishes': tp.values})

stat_data = pd.merge(df, df2)
```

```python
stat_data['Performance ratio percentage'] = stat_data['Top Finishes']/stat_data['Total Finishes']*100
```

# Data analysis

I've settled on >5 wins as a reasonable cutoff point for such a small event (only 8 rounds total into a top 8). I did this partly because just top 8 was too narrow and too skewed. Because of the above average skill level at a Mythic Championship, any >5 win deck in my opinion could have made top 8, depending on matchups and piloting skill.

```python

plt.xticks(rotation=90)
plt.plot(stat_data['Deck'], stat_data['Total Finishes'], label='Total decks')
plt.plot(stat_data['Deck'], stat_data['Performance ratio percentage'], label='>5 wins conversion rate')
plt.legend(fontsize='x-large')

```

### Conclusion
Pretty clearly, the most popular deck was Bant Golos, but it's also very low in terms of mean wins.

This is likely the result of the deck being "hated out" of the format; people were playing anti-meta decks specifically to beat it. Maindeck Disdainful Stroke was a common occurrence because of this. The event being won by Gruul Aggro, a deck that according to the pilot is built to beat Bant Golos, is another piece of evidence for the anti-meta field.

One very clear outlier was Golgari Adventure, which heavily underperformed while being picked up by historically strong players (like Piotr Glogowski and Luis Salvatto).

For further investigation, more grouping can be done, as 4c Golos plays almost the same as Bant Golos, and Simic Food & Bant Food also play very similarly.
